###### Task Description: 
    We need to read the image in RBG format and then convert it to a grayscale image. This will turn an image into a classic black and white photo. Then the next thing to do is invert the grayscale image also called negative image, this will be our inverted grayscale image. Inversion can be used to enhance details. Then we can finally create the pencil sketch by mixing the grayscale image with the inverted blurry image. This can be done by dividing the grayscale image by the inverted blurry image.

#### Importing The Libraries

```python
import cv2
import numpy as np
from PIL import Image
from IPython.display import display
```

```python
original_img = cv2.imread(r'C:\Users\shash\OneDrive\Desktop\LETS GROW MORE\TASK 2\test_img.jpg')
```

```python
original_img = cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB)
```

```python
display(Image.fromarray(original_img))
```

#### Converting the Original Image into Grayscale

```python
gray_image = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

#### Inverting the Grayscale image

```python
inv_gray_image = 255 - gray_image
```

```python
inv_gray_image
```

```python
display(Image.fromarray(inv_gray_image))
```

#### Blur the image

```python
blur_image = cv2.GaussianBlur(inv_gray_image, (35, 35), 0)
```

```python
display(Image.fromarray(blur_image))
```

#### Mixing the grayscale image and inverted blur image
